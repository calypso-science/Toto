```python
import pandas as pd
import toto
from toto.inputs.mat import MATfile
import matplotlib.pyplot as plt
import datetime

filename=r'../_tests/mat_file/tidal.mat'
tx=MATfile(filename)
df=tx._toDataFrame()

dfout=df[0].TideAnalysis.detide(mag='el_tide',\
        args={'Minimum SNR':2,\
        'Latitude':-36.0,
        'folder out':'/tmp/'
        })


fig, ax = plt.subplots(1)
ax.plot(df[0].index,df[0]['el_tide'],label='Observed')
ax.plot(dfout.index,dfout['el_tidet'],'r',label='Predicted')
plt.xlim(datetime.datetime(2010,11,1),datetime.datetime(2010,12,1))
plt.legend()
fig.autofmt_xdate()
plt.show()




dfout=df[0].TideAnalysis.tidal_stat(mag='el_tide',\
        args={'Minimum SNR':2,\
        'Latitude':-36.0,
        'folder out':'/tmp/'
        })

stat = pd.read_excel('/tmp/Concstats.xlsx')
stat.head()
```
